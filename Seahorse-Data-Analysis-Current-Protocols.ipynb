```python
#The following script can be used to curate data for basal OCR and mito OCR
#Import all required libraries
import pandas as pd

#Define the number of replicate wells
rep = int(6)

#Import csv data file (in the same folder/directory as this notebook)
df=pd.read_csv('Sample_Normalized_OCR_Data.csv')

#Define basal_OCR list and OCR_RotAA list
df_basal_OCR = df[0:3]
df_OCR_RotAA = df[6:9]

#Calculate mean for basal_OCR
basal_meanSD = df_basal_OCR.describe()
basal_OCR = basal_meanSD.T['mean']

#Calculate data distribution for each treatment groups
basal_OCR_1 = basal_OCR[0:rep].describe()
basal_OCR_2 = basal_OCR[rep:2*rep].describe()
basal_OCR_PB = basal_OCR[2*rep:3*rep].describe()
basal_OCR_Neg = basal_OCR[3*rep:(4*rep+2)].describe()

#Combine basal OCR data into one dataframe
glued_basal_OCR = pd.concat([basal_OCR_1, basal_OCR_2, basal_OCR_PB], axis = 1)
#Calculate fold changes of the basal OCR for each treatment groups
FC_basal_OCR = glued_basal_OCR.T['mean']/basal_OCR_Neg.T['mean']
FC_basal_OCR = FC_basal_OCR.rename({'mean':'basal_OCR_FC'})

#Calculate data distribution for OCR_RotAA
RotAA_meanSD = df_OCR_RotAA.describe()

#Calculate mito OCR by substracting mean of OCR_RotAA from basal_OCR 
mito_OCR = basal_meanSD.T['mean'] - RotAA_meanSD.T['mean']

#Calculate data distribution for each treatment groups
mito_OCR_1 = mito_OCR[0:rep].describe()
mito_OCR_2 = mito_OCR[rep:2*rep].describe()
mito_OCR_PB = mito_OCR[2*rep:3*rep].describe()
mito_OCR_Neg = mito_OCR[3*rep:(4*rep+2)].describe()

#Combine mito OCR data into one dataframe
glued_OCR = pd.concat([mito_OCR_1, mito_OCR_2, mito_OCR_PB], axis = 1)
#Calculate fold changes of the mito OCR for each treatment groups
FC_mito_OCR = glued_OCR.T['mean']/mito_OCR_Neg.T['mean']
FC_mito_OCR = FC_mito_OCR.rename({'mean':'mito_OCR_FC'})

#Combine fold changes data of basal OCR and mito OCR into one dataframe
FC_basal_OCR = FC_basal_OCR.reset_index()
FC_mito_OCR = FC_mito_OCR.reset_index()
glued = pd.concat([FC_basal_OCR, FC_mito_OCR], axis = 1)

#Export fold changes data to excel
glued.to_excel(excel_writer = 'Result_Basal_Mito_OCR.xlsx',sheet_name='FC', 
               na_rep='', float_format=None, columns=None, header=True, 
               index=True, index_label=None, startrow=0, startcol=0, 
               engine=None, merge_cells=True, encoding=None, inf_rep='inf', 
               verbose=True, freeze_panes=None)

#Calculate upper and lower 2SD limit of the negative control in fold change
UpperFC_mito_OCR = (mito_OCR_Neg.T['mean']+ (2*mito_OCR_Neg.T['std']))/mito_OCR_Neg.T['mean']
LowerFC_mito_OCR = (mito_OCR_Neg.T['mean']- (2*mito_OCR_Neg.T['std']))/mito_OCR_Neg.T['mean']
UpperFC_basal_OCR = (basal_OCR_Neg.T['mean']+ (2*basal_OCR_Neg.T['std']))/basal_OCR_Neg.T['mean']
LowerFC_basal_OCR = (basal_OCR_Neg.T['mean']- (2*basal_OCR_Neg.T['std']))/basal_OCR_Neg.T['mean']
UpperFC_mito_OCR
LowerFC_mito_OCR
UpperFC_basal_OCR
LowerFC_basal_OCR
```
